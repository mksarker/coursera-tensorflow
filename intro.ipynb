# Import libraries

```
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape = [1])]) # define the model
model.compile(optimizer = 'sgd', loss = 'mean_squared_error') # loss function and optimizer
xs = np.array([-1.0 , 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float) # data for x
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float) # data for y
model.fit(xs, ys, epochs = 500) # train
```

```
print(model.predict([10.0])) # predict
```
